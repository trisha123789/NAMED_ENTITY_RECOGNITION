```python
import spacy
```

```python
nlp = spacy.load("en_core_web_sm")
```

```python
nlp.pipe_names
```

```python
text = nlp("Tesla is aquiring twitter for $45 billion")
```

```python
for ent in text.ents:
  print(f"{ent.text} | {ent.label_} | {spacy.explain(ent.label_)}")
```

```python
text = nlp("Tesla is aquiring twitter for forty five billion")
```

```python
for ent in text.ents:
  print(f"{ent.text} | {ent.label_} | {spacy.explain(ent.label_)}")
```
